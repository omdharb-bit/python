```python
# if statement

age = int(input("Enter ur age:"))

if age >= 18:
    print("You can cast your vote")

# else statement
else:
    print("You cannot cast your vote")
```

```python
# elif statement

age = int(input("Enter your age: "))

if age < 13:
    print("You are a child")

elif age < 18:
    print("You are a teenager")

else:
    print("You are an adult")
```

```python
# Nested Conditional statemnt

num = int(input("Enter a number: "))

if num > 0:
    print("No. is positive")
if num < 0:
    print("No. is negative")
if num % 2 == 0:
    print("No. is even")
else:
    print("No. is odd")
```

```python
# Q- Determine if a year is leap year or not usinng nested loop

year = int(input("Enter year you want to check that it is leap year or not: "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(year, "It is a leap year")
        else:
            print(year, "It is not a leap year")
    else:
        print(year, "It is a leap year")
else:
    print(year, "It is not a leap year")
```
