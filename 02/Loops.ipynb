```python
# for loop

for i in range(5):
  print(i)
 
```

```python
# for i in range(1,10,1):
for i in range(1,10,2):
  print(i)
```

```python
for i in range(10,1,-3):     #middle value means skiping that no. and -3 means jump three no. backwords
  print(i)
```

```python
# whike loop

count=0

while count<5:
  print(count)
  count+=1
```

```python
# loop control statement

# break 

for i in range(10):
  if i==5:
    break
  print(i)
  
   
```

```python
# continue skips the current iteration

for i in range(10):
  if i %2==0:
    continue
  print (i)
```

```python
# pass
# it is null operation it does nothing

for i in range(5):
  if i==3:
    pass
  print(i)
```

```python
# Nested  loop
# loop inside a loop

for i in range(3):
  for j in range(2):
    print(f"i:{i} and j:{j}")
```

```python
# examples

#  Calculate the sum of first N natural numbers usinng a while and for loop

# while loop

n=int(input("Enter the no. till the value you want sum: "))
sum=0
count=1

while count<=n:
  sum+=count
  count=count+1
  

  print("Sum of first entered natural no: ",sum)
```

```python
n=10
sum=0

for i in range (n+1):
  sum+=i
  print(sum)
```
